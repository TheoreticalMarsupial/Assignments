```python
# stacked band 1: nir, red, green image

import numpy as np
import rasterio

def scale(band): # scale values for display purposes
    return band / 50.0

image_filename = "old_town.tif"
my_image = rasterio.open(image_filename)

blue = scale(my_image.read(1))
green = scale(my_image.read(2))
red = scale(my_image.read(3))
nir = scale(my_image.read(4))

# stacked band 1: nir, red, green
nrg = np.dstack((nir,red,green))
nrg.shape

# reorder stacked band 1
nrg_reordered = np.moveaxis(nrg, [0, 1, 2], [1, 2, 0])

# Check the shape. 
nrg_reordered.shape 

nrg_reordered = nrg_reordered.astype('uint8')
nrg_reordered.dtype

with rasterio.open(
    'stackedband1_nrg.jpg',                              
    'w',                                            
    driver='JPEG',                                  
    compress='JPEG',
    height=nrg_reordered.shape[1],                  
    width=nrg_reordered.shape[2],                   
    count=nrg_reordered.shape[0],                   
    dtype=nrg_reordered.dtype,                      
    crs=my_image.profile['crs'],                    
    transform=my_image.profile['transform']         
    ) as my_raster_writer:
        my_raster_writer.write(nrg_reordered)       
        
        
        
# stacked band 2: red, nir, blue image

def scale(band): # scale values for display purposes
    return band / 50.0

image_filename = "old_town.tif"
my_image = rasterio.open(image_filename)

blue = scale(my_image.read(1))
green = scale(my_image.read(2))
red = scale(my_image.read(3))
nir = scale(my_image.read(4))

# stacked band 2: red, nir, blue
rnb = np.dstack((red,nir,blue)) 
rnb.shape

# reorder stacked band 2
rnb_reordered = np.moveaxis (rnb, [0,1,2], [1,2,0])
rnb_reordered.shape

rnb_reordered = rnb_reordered.astype('uint8')
rnb_reordered.dtype

with rasterio.open(
    'stackedband2_rnb.jpg',                               
    'w',                                            
    driver='JPEG',                                 
    compress='JPEG', 
    height=rnb_reordered.shape[1],                  
    width=rnb_reordered.shape[2],                  
    count=rnb_reordered.shape[0],                   
    dtype=rnb_reordered.dtype,                     
    crs=my_image.profile['crs'],                    
    transform=my_image.profile['transform']         
    ) as my_raster_writer:
        my_raster_writer.write(rnb_reordered)       
        
        
# stacked band 3: nir, blue, red image

def scale(band): # scale values for display purposes
    return band / 50.0

image_filename = "old_town.tif"
my_image = rasterio.open(image_filename)

blue = scale(my_image.read(1))
green = scale(my_image.read(2))
red = scale(my_image.read(3))
nir = scale(my_image.read(4))

# stacked band 2: red, nir, blue
nbr = np.dstack((nir,blue,red)) 
nbr.shape

# reorder stacked band 2
nbr_reordered = np.moveaxis (nbr, [0,1,2], [1,2,0])
nbr_reordered.shape

nbr_reordered = nbr_reordered.astype('uint8')
nbr_reordered.dtype

with rasterio.open(
    'stackedband3_nbr.jpg',                               
    'w',                                            
    driver='JPEG',                                  
    compress='JPEG', 
    height=nbr_reordered.shape[1],                 
    width=nbr_reordered.shape[2],                  
    count=nbr_reordered.shape[0],                   
    dtype=nbr_reordered.dtype,                      
    crs=my_image.profile['crs'],                    
    transform=my_image.profile['transform']         
    ) as my_raster_writer:
        my_raster_writer.write(nbr_reordered)       
```
