```python
import rasterio
from matplotlib import pyplot as plt
from rasterio.plot import show

image_file = "T40DEL_20180110T051711_TCI.jp2"

my_imageTC = rasterio.open(image_file)
my_imageTC

show(my_imageTC)
```

```python
# Present number of image bands 
print(my_imageTC.count)

# Present number of indexes
print(my_imageTC.indexes)
```

```python
# Band 8 - NIR
image_fileB8 = "T40DEL_20180110T051711_B08.jp2"

my_imageB8 = rasterio.open(image_fileB8)
my_imageB8

show(my_imageB8)
```

```python
# Present number of image bands 
print(my_imageB8.count)

# Present number of indexes
print(my_imageB8.indexes)
```

```python
# Band 4 - Red
image_fileB4 = "T40DEL_20180110T051711_B04.jp2"

my_imageB4 = rasterio.open(image_fileB4)
my_imageB4

show(my_imageB4)
```

```python
# Present number of image bands 
print(my_imageB4.count)

# Present number of indexes
print(my_imageB4.indexes)
```

```python
# Band 3 - Green
image_fileB3 = "T40DEL_20180110T051711_B03.jp2"

my_imageB3 = rasterio.open(image_fileB3)
my_imageB3

show(my_imageB3)
```

```python
green = my_imageB3.read()
```

```python
green
```

```python
import numpy as np

def scale(band): # scale values for display purposes
    return band / 1000.0

nir = scale(my_imageB8.read(1))
red = scale(my_imageB4.read(1))
green = scale(my_imageB3.read(1))

nrg= np.dstack((nir, red, green))                     # stack of all three bands 
nrg
```
