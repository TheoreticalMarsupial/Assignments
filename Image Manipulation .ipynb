```python
# Band 8 - NIR
image_file = "T40DEL_20180110T051711_B08.jp2"

my_imageB8 = rasterio.open(image_file)
my_imageB8

from rasterio.plot import show
show(my_imageB8)
```

```python
# Band 3 - Green
image_file = "T40DEL_20180110T051711_B03.jp2"

my_imageG3 = rasterio.open(image_file)
my_imageG3

from rasterio.plot import show
show(my_imageG3)
```

```python
# Band 2 - Blue
image_file = "T40DEL_20180110T051711_B02.jp2"

my_imageB2 = rasterio.open(image_file)
my_imageB2

from rasterio.plot import show
show(my_imageB2)
```

```python
nir = my_imageB8.read(1)
green = my_imageG3.read(1)
blue = my_imageB2.read(1)
```

```python
import numpy as np

def scale(band): # scale values for display purposes
    return band / 1000.0

nir = scale(my_imageB8.read(1))
green = scale(my_imageG3.read(1))
blue = scale(my_imageB2.read(1))



ngb = np.dstack((nir, green, blue))                     # stack of all three bands 
ngb
```
